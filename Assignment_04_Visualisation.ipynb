**4. Using the given x and y data, generate a 3D surface plot to visualize the function image.png**

image.png

```python
#Creation of the Dataset

x = 0.25*np.linspace(-5, 5, 100)
y = 0.25*np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
df.sample(5)
#df.info()
```

```python
#Creating the 3D Surface

surface=go.Surface(x=df['X'],y=df['Y'],z=df['Z'],colorscale='Inferno',showscale=True)

fig=go.Figure(data=[surface])
fig.show()
```

# **04. BOKEH**

**Q1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x**

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")
p.line(x, y, line_width=2, line_color="blue", legend_label="Sine Wave")

# Display the plot
show(p)
```

**Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.**

* fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
* counts = [20, 25, 30, 35]

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label="Fruits", y_axis_label="Count")
p.vbar(x=fruits, top=counts, legend_label="fruits", width=0.5, bottom=0, color=["red", "orange", "yellow", "green"])
show(p)
```

**Q4. Create a Bokeh histogram to visualize the distribution of the given data.**

* data_hist = np.random.randn(1000)
* hist, edges = np.histogram(data_hist, bins=30)

```python
data_hist = np.random.randn(1000)  # Assuming your data is in data_hist
hist, edges = np.histogram(data_hist, bins=30)

# Calculate centers of bins for plotting
centers = (edges[:-1] + edges[1:]) / 2

# Create a new plot with a title and axis labels
p = figure(title="Histogram of Random Data", x_axis_label="Value", y_axis_label="Count")
p.quad(left=centers, right=edges[1:], top=hist, bottom=0, line_color="gray", fill_color="skyblue", legend_label="Histogram")
show(p)
```
